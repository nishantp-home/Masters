```python
# Bar plots from aggregated values
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names']+['species'])
iris.species = np.where(iris.species == 0.0, 'setosa', np.where(iris.species == 1.0, 'versicolor', 'virginica'))
iris.columns = iris.columns.str.replace(' ', '') # Remove empty spaces 
iris.columns
```

```python
iris.head(6)
```

```python
i = iris.mean()
i
```

```python
my_plot = i.plot(kind='bar')
```

```python
isa = iris.groupby(['species']).mean()

myplot = isa.plot(kind='bar')
myplot.set_ylabel('Average values')
```

```python
sd = iris.groupby(['species']).std()

isa.plot(kind='bar', yerr=sd)
```
